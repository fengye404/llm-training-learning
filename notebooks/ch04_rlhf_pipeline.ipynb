```python
from pathlib import Path
import subprocess, sys, re

ROOT = Path.cwd()
if not (ROOT / 'projects').exists():
    if (ROOT.parent / 'projects').exists():
        ROOT = ROOT.parent
    else:
        raise RuntimeError(f'未找到项目根目录（应包含 projects/ 目录），当前目录: {Path.cwd()}')

print('使用项目目录:', ROOT)

def run_py(rel_path: str) -> str:
    # 执行项目脚本并返回标准输出文本
    result = subprocess.run(
        [sys.executable, str(ROOT / rel_path)],
        cwd=ROOT,
        capture_output=True,
        text=True,
        check=True,
    )
    print(result.stdout)
    return result.stdout
```

## 运行 RLHF 教学脚本


```python
out = run_py('projects/project-03-rlhf-pipeline/rlhf_pipeline_demo.py')
```

## 自动解读关键信号


```python
reward_values = [float(x) for x in re.findall(r'平均期望奖励=([0-9.]+)', out)]
print('期望奖励序列:', reward_values)
print('是否上升:', reward_values[-1] > reward_values[0])

best_lines = [line.strip() for line in out.splitlines() if line.strip().startswith('最优策略=')]
print('\n最终最优策略:')
for line in best_lines:
    print(line)
```

## 常见误区

- 奖励上升 != 绝对正确，只表示“更符合当前奖励函数”。
- 奖励函数设计不合理，模型也可能学偏。
