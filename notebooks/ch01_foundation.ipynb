# 第1章课堂版：训练最小闭环（零基础）

这章只解决一个问题：**模型为什么会越练越准**。

你不需要先懂 AI 数学，只要盯住 3 个信号：
1. `loss` 是否下降（错误是否减少）
2. `w` 是否接近 2、`b` 是否接近 1（参数是否学对）
3. 测试样例 `x=3` 的预测是否接近 7


## 学习目标（本章过关标准）

- 你能用白话解释：训练 = 反复纠错更新参数。
- 你能解释 `loss` 为什么应该下降。
- 你能改一个参数（如学习率）并预测变化方向。


```python
from pathlib import Path
import subprocess, sys, re

ROOT = Path.cwd()
if not (ROOT / 'projects').exists():
    if (ROOT.parent / 'projects').exists():
        ROOT = ROOT.parent
    else:
        raise RuntimeError(f'未找到项目根目录（应包含 projects/ 目录），当前目录: {Path.cwd()}')

print('使用项目目录:', ROOT)

def run_py(rel_path: str) -> str:
    # 执行项目脚本并返回标准输出文本
    result = subprocess.run(
        [sys.executable, str(ROOT / rel_path)],
        cwd=ROOT,
        capture_output=True,
        text=True,
        check=True,
    )
    print(result.stdout)
    return result.stdout
```

## 第一步：运行第1章脚本

运行后你会看到多行日志，每行都是某个 epoch 的训练状态。


```python
out = run_py('projects/project-00-foundation/toy_autograd_train.py')
```

## 第二步：自动解读输出（避免“看不懂日志”）

下面这个单元会从日志中提取关键字段，帮你判断是否学到了。


```python
loss_values = [float(x) for x in re.findall(r'loss=([0-9.]+)', out)]
w_values = [float(x) for x in re.findall(r'w=([0-9.]+)', out)]
b_values = [float(x) for x in re.findall(r'b=([0-9.]+)', out)]

print('loss 记录点数:', len(loss_values))
print('首个 loss:', loss_values[0])
print('最后 loss:', loss_values[-1])
print('loss 是否下降:', loss_values[-1] < loss_values[0])
print('最终 w:', w_values[-1], '(目标约 2)')
print('最终 b:', b_values[-1], '(目标约 1)')
```

## 第三步：你应该如何解释这个结果

- 如果 `loss` 下降：说明模型在减少错误。
- 如果 `w,b` 接近目标：说明参数在朝正确方向收敛。
- 如果最终测试值接近预期：说明模型具备基本泛化能力。

> 这就是训练的本质：参数被数据“推着走”。
